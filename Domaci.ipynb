Zakljucujemo da 0% uzoraka fali. Sledece sto cemo odrediti jeste koliko ima nedostajucih vrednosti za prosecnu brzinu vetra.

```python
print('Brzin vetra:\n', df['Iws'].unique())
print('Iprec:\n', df['Iprec'].unique())
print('PM2.5:\n', df['PM_US Post'].unique())
```

```python
df.loc[df['year']==2010, 'year']=np.nan
df.drop(df.loc[df['Iws']>14].index, axis=0, inplace=True)
df.head()
```

Izbacio sam 2010. godinu, jer u njoj nemamo merenu vrednost ni za jednu PM cesticu. Takodje, za brzinu vetra preko 8m/s sam stavio vrednosti NaN.

```python
df['precipitation'].fillna(method='ffill', inplace=True)
df['Iprec'].fillna(method='ffill', inplace=True)
df['PM_US Post'].fillna(df['PM_US Post'].median(), inplace=True)
df['Iws'].fillna(df['Iws'].median(), inplace=True)
df.describe()
```
